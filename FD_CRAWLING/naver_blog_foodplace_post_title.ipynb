```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import openpyxl
import datetime
import time
import pandas
```

### 동적 웹 크롤링 : 맛집 블로그 포스트 100개 크롤링 - 제목만

```python
url = "https://section.blog.naver.com/ThemePost.naver?directoryNo=29&activeDirectorySeq=3&currentPage=1"
driver = webdriver.Chrome()
driver.get(url)
time.sleep(3)
```

```python
post_titles = driver.find_elements(By.CLASS_NAME,'title_post')

crawler_time = datetime.datetime.now().strftime('%Y-%m-%d %H:%M')
print(crawler_time)
App = "Naver Blog"
data = []

print("-----네이버 블로그 - 맛집 포스트 제목-----")
for page in range(1,3): ##1~2페이지까지
    driver.find_element(By.XPATH, '/html/body/ui-view/div/main/div[1]/div/section/div[2]/div[11]/span[{}]/a'.format(page)).click()
    time.sleep(2)

    post_titles = driver.find_elements(By.CLASS_NAME,'title_post')

    for post_title in post_titles[9:]: ## Top 3 제거(제목너무 짧음), 공백 행 제거
        title = post_title.text
        line_data = []
        line_data.append(App)
        line_data.append(title)
        line_data.append(crawler_time)
        # print(line_data) # 한 행 씩 출력
        data.append(line_data)

        # print(title)

    # print(data)
df = pandas.DataFrame(data, columns= ['App', 'Post Title', 'Crawler Time'])
df.index = df.index+1
df
```

```python
df.to_excel('FD_Crawling_Naver_Blog.xlsx')
```
